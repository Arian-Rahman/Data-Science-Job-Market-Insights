```python

def update_ratings(df, rating_col, *cols):
    """
    Update the 'rating' column in the DataFrame with numeric values from specified columns.

    Parameters:
    - df: pandas DataFrame containing the data.
    - rating_col: Name of the column where ratings will be stored.
    - *cols: Columns to extract numeric values from.

    Returns:
    - DataFrame with updated 'rating' column.
    """
    # Initialize the rating column with numeric values from the first specified column
    df[rating_col] = pd.to_numeric(df[cols[0]], errors='coerce')

    # Fill NaN values in 'rating' column with numeric values from the other columns that are convertable into numeric data
    for col in cols[1:]:
        df[rating_col] = df[rating_col].fillna(pd.to_numeric(df[col], errors='coerce'))

    return df

# col 1,2,3,4 has rating values scattered
split_columns = update_ratings(split_columns, 'rating', 1,2,3, 4)

split_columns['rating'].info()
```

```python

def update_company_names(df, name_col, col, exclude_values):
    """
    Update the 'name_col' column with filtered values from 'col', excluding those in 'exclude_values'.

    Parameters:
    - df: DataFrame
    - name_col: Name of the column to update.
    - col: Column from which to filter values.
    - exclude_values: List of values to exclude.

    Returns:
    - DataFrame with updated 'name_col' column.
    """
    # Create a mask for values to include , meaning those , that are not in exclude_values
    mask = ~df[col].isin(exclude_values)

    # Create a new column with the filtered values where the mask is True
    df[name_col] = df[col].where(mask, other=None)

    return df

exclude_values_list = ['Remote', '3.2', '3.8', 'Falls Church, VA•Remote', '4.0', '3.4', 'Hybrid work', '3.9', 'San Francisco, CA•Remote', '3.3']

split_columns = update_company_names(split_columns, 'company_name', 2, exclude_values_list)

# Verify the result
print(split_columns["company_name"].unique())
```

# 4.1 Salary column data engineering

```python
job_info_backup_v3 = job_info.copy()

job_info['salary']
```

```python

def is_numeric(value):
    try:
        float(value.replace(',', '').replace('$', ''))
        return True
    except ValueError:
        return False

# Function to extract min and max salary
def extract_salaries(salary):
    parts = salary.split()
    min_salary, max_salary = None, None

    for i, part in enumerate(parts):
        # Check if the part is numeric (after removing $, commas)
        if is_numeric(part):
            value = float(part.replace(',', '').replace('$', ''))
            if min_salary is None:
                min_salary = value
            elif max_salary is None:
                max_salary = value
                break
    return min_salary, max_salary

def extract_salary_type(salary):
    if 'year' in salary:
        return 'per year'
    elif 'day' in salary:
        return 'per day'
    elif 'hour' in salary:
        return 'per hour'
    else:
        return None

job_info['salary_type'] = job_info['salary'].apply(extract_salary_type)

job_info[['min_salary', 'max_salary']] = job_info['salary'].apply(lambda x: pd.Series(extract_salaries(x)))
#job_info['salary_type'] = job_info['salary'].apply(extract_salary_type)
print(job_info[['salary', 'min_salary', 'max_salary', 'salary_type']])
```

```python
job_info.head()
```

```python
# calculate average first ,row by row
def average_calculator(row):
  if pd.notna(row['max_salary']):
     avg_salary = (row['min_salary'] + row['max_salary']) / 2
     return avg_salary
  else :
    avg_salary = row['min_salary']
    return avg_salary

job_info['average_salary'] = job_info.apply(average_calculator,axis=1)

print(job_info[['salary', 'min_salary', 'max_salary', 'salary_type', 'average_salary']])
```

```python

# Constants for conversion
WORKDAYS_PER_YEAR = 260
HOURS_PER_DAY = 8

def calculate_standardized_salary(row):
    '''
    Function to calculate standardized yearly salary
    '''
    salary_type = row['salary_type']

    if salary_type is None:
        return None

    if 'year' in salary_type:
        return row['average_salary']
    elif 'day' in salary_type:
        return row['average_salary'] * WORKDAYS_PER_YEAR
    elif 'hour' in salary_type:
        return row['average_salary'] * HOURS_PER_DAY * WORKDAYS_PER_YEAR
    else:
        return None

job_info['yearly_salary'] = job_info.apply(calculate_standardized_salary, axis=1)

job_info[['salary', 'min_salary', 'max_salary', 'salary_type','average_salary', 'yearly_salary']]
```

```python
ranges = [
    (10000, 50000),
    (50001, 100000),
    (100001, 150000),
    (150001, 200000),
    (200001, 250000),
    (250001, 300000),
    (300001, 400000),
    (400001, 500000),
    (500001, float('inf'))
]
labels = range(1, len(ranges) + 1)

# Function to assign categories and ranges based on salary
def assign_category_and_range(salary):
    for (lower, upper), label in zip(ranges, labels):
        if lower <= salary <= upper:
            return label, f"${lower} - ${upper}"
    return None, None

job_info[['Category', 'Range']] = job_info['yearly_salary'].apply(assign_category_and_range).apply(pd.Series)
# Using pd.Series to split a tuple returned by the function into separate columns in the DF


print(job_info[['yearly_salary', 'Category', 'Range']])
```
